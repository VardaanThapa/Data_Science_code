```python
plt.subplot(2,2,1)
plt.scatter(tips.index,tips.tip)
```

```python
plt.subplot(2,2,2)
sns.stripplot(data = tips ,y='tip',x='day',jitter=True)
plt.subplot(2,2,3)
sns.swarmplot(data=tips,x='tip',y='day',hue="sex",orient='h')
plt.show()
```

```python
plt.subplot(2,2,1)
sns.swarmplot(data=tips,x='tip',y='day',hue="sex",orient='h')
```

```python
sns.boxplot(data=tips,y='tip',x='day',hue='sex')
plt.ylabel("tips$")
```

```python
sns.violinplot(data=tips,y='tip',x='day')
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip',kind='kde')
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
```

```python
data = pd.read_csv("train.csv")
```

```python
data.head()
```

```python
data.dtypes
```

```python
sns.barplot(data=data,y='Survived',x='Sex',hue='Pclass',estimator=sum)
```

```python
grid = sns.FacetGrid(data=data, col = "Survived", row='Pclass', size=2.4,aspect=1.5)
grid.map(plt.hist,'Age')
grid.add_legend()
```

```python
sns.distplot(a=data.Age, kde=True, bins=3)
```

```python
fig,axes = plt.subplots(2,4,figsize=(16,10))
sns.distplot(a=data.Age, kde=True, bins=3,ax=axes[1,1])
plt.savefig("image.jpg")
```
